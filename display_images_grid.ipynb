```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Set the folder containing images
image_folder = 'output/lbp-faces/'

# List all files in the directory
images = [img for img in os.listdir(image_folder) if img.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

# Number of images per row
images_per_row = 4

# Number of rows
num_rows = len(images) // images_per_row + (1 if len(images) % images_per_row != 0 else 0)

# Create a figure with the specified number of subplots
fig, axs = plt.subplots(num_rows, images_per_row, figsize=(15, 3 * num_rows))

# Flatten the axis array for easy indexing
axs = axs.flatten()

# Display images in a grid
for i, img_name in enumerate(images):
    img_path = os.path.join(image_folder, img_name)
    img = mpimg.imread(img_path)
    axs[i].imshow(img)
    axs[i].set_title(img_name)
    axs[i].axis('off')  # Hide axes

# Hide any unused subplots
for j in range(i + 1, len(axs)):
    axs[j].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
